```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("../Data/train.csv")
```

```python
train = train.drop("id", axis = 1)
train
```

```python
defect_counts = train[["Pastry","Z_Scratch","K_Scatch","Stains","Dirtiness","Bumps","Other_Faults"]].sum(axis=1)
train = train[defect_counts==1]
```

```python
train['target'] = train[["Pastry","Z_Scratch","K_Scatch","Stains","Dirtiness","Bumps","Other_Faults"]].apply(lambda row: row.idxmax(), axis=1)
train = train.drop(["Pastry","Z_Scratch","K_Scatch","Stains","Dirtiness","Bumps","Other_Faults"],axis = 1)
```

```python
label_map = {'Pastry': 0, 'Z_Scratch': 1, 'K_Scatch': 2, 'Stains': 3, 'Dirtiness': 4, 'Bumps': 5, 'Other_Faults': 6}

train['target'] = train['target'].map(label_map)
```

```python
train
```

```python
from pycaret.classification import *
s = setup(train, target = 'target', session_id = 123)
```

```python
best = compare_models()
```

```python
predict_model(best, raw_score=True)
```

```python
test = pd.read_csv("../Data/test.csv")
```

```python
predictions = predict_model(best, data = test.drop("id", axis=1),raw_score=True)
```

```python
predictions
```

```python
testid = test["id"]
testpred = predictions[["prediction_score_0","prediction_score_1","prediction_score_2","prediction_score_3","prediction_score_4","prediction_score_5","prediction_score_6"]]
```

```python
submitFile = testpred
submitFile["id"] = testid
submitFile
```

```python
last_column = submitFile.iloc[:, -1]

submitFile = submitFile.iloc[:, :-1]

submitFile = pd.concat([last_column, submitFile], axis=1)
submitFile
```

```python
submitFile.to_csv('../Submissions/BasedModelSub.csv', index=False)
```
